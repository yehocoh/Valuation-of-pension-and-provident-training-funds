```python
# SAMPLE FILE לגיליון מניות
import openpyxl, os,  pandas as pd

def process_workbook(parameter1,sheet):
    file_name = os.path.basename(parameter1)
    workbook = openpyxl.load_workbook(parameter1, read_only=True, data_only=True)
    sheet = workbook[sheet]
    data = sheet.values
    # headers = next(data)[13:]
    rows = list(data)
    df = pd.DataFrame(rows)#, columns=headers)
    df = df.iloc[13:,:15]
    df.insert(0, 'File Name', file_name)
    return df
```

```python
# ALL FILES לגיליון מניות

import pandas as pd
import os

def process_folder(folder_path,sheet):
    # Get a list of all the files in the folder
    files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    # Filter out hidden files
    files = [f for f in files if not f.startswith('.')]

    # Define a list to store the transformed data
    data = []

    # Iterate through the files and apply the transformation function to each one
    for file in files:
        file_path = os.path.join(folder_path, file)
        df = process_workbook(file_path,sheet)
        df['Source.Name'] = file
        df['comp_num'] = file[:9]
        df['QTR'] = file.split("_")[2][1:2]
        df['YEAR'] = file.split("_")[2][2:4]
        df['Maslul'] = file.split("_")[1]
        df['sheet'] = sheet
        data.append(df)

    # Concatenate all the data frames into a single data frame
    df = pd.concat(data)

    # Rename the columns
    df = df.rename(columns={1: 'NAME',
                            2: 'ISIN',
                            3: 'EXC',
                            4: 'INFO_PROVIDER',
                            5: '#ISU',
                            6: 'SEGMENT',
                            7: 'CUR',
                            8: 'UNIT',
                            9: 'Price_0',
                            10: 'EXT',
                            11: 'PV',
                            12: '%2',
                            13: '%1',
                            14: '%3'})
    df = df.drop(columns=0)

    # Convert the data types of the columns
    df = df.astype({'Price_0': 'float',
                    'UNIT': 'float',
                    'EXT': 'float',
                    'PV': 'float',
                    '%1': 'float',
                    '%2': 'float',
                    '%3': 'float'})

    # Filter out rows with missing or empty values in the ISIN column
    df = df[df['ISIN'].notnull() & (df['ISIN'] != '')]

    # filter the DataFrame and keep only the rows where the value in the "ISIN" column has a length of 12
    df =  df.loc[df["ISIN"].apply(lambda x: len(x) == 12)]

    return df
```

```python
# שליפת הציטוטים לפי ISIN
import yfinance as yf

# Create an empty list to store the stock prices
prices = {}

# Iterate over the ISINs in the DataFrame
for isin in ISIN:
    # Try to get the stock information using yfinance
    try:
        stock_info = yf.Ticker(isin).info
        # Append the stock price to the list
        prices[isin] = stock_info["regularMarketPreviousClose"]
    except :
        # If the stock is not found, append a null value to the list
        print(f"Unable to retrieve stock price for ISIN {isin} - - {stock_info['shortName']}")
        prices[isin] = (None)
```

```python
# ריצה נוספת למקרה שהחיבור לאיפיאיי לא עבד כיאות - רק על נתוני חסרים
# Iterate over the ISINs in the DataFrame
for isin in prices.keys():
  if prices[isin] == None:
    # Try to get the stock information using yfinance
    try:
        stock_info = yf.Ticker(isin).info
        # Append the stock price to the list
        prices[isin] = stock_info["regularMarketPreviousClose"]
    except :
        # If the stock is not found, append a null value to the list
        print(f"Unable to retrieve stock price for ISIN {isin}")
        prices[isin] = (None)

# Add the stock prices to the DataFrame
prices
```

```python
# Convert the dictionary to a Pandas Series
isin_price_series = pd.Series(prices)

# Convert the Series to a DataFrame
isin_price_df = isin_price_series.to_frame()

# Rename the column to match the column name in the table
isin_price_df.rename(columns={0: "Price"}, inplace=True)

# Merge the table and the dictionary based on the "ISIN" column
merged_table = pd.merge(df, isin_price_df, left_on="ISIN", right_index=True)
```

```python
merged_table.head(1)
```

```python
# הורדת ני"ע שאין להם מחיר וני"ע שאינם דולריים (לצורך הצגה)
merged_table = merged_table[merged_table["Price"]!= None]
merged_table = merged_table[merged_table["CUR"]=="דולר אמריקאי"]
```

```python
# ייבוא שע"ח דולד שקל עדכני לחישוב שווי נוכחי
rate = yf.Ticker("USDILS=X").info["regularMarketPreviousClose"]
print(rate)

# חישוב שווי נוכחי של התיק ופערים בין השווי המדווח למעודכן
merged_table['new_pv'] = (merged_table['UNIT']*merged_table['Price']*rate/1000).astype(float)
merged_table['gap#'] = (-merged_table['PV']+merged_table['new_pv']).astype(float)
merged_table['gap%'] = (merged_table['gap#']/merged_table['PV']).astype(float)
merged_table['OLD_PRICE'] = merged_table['Price_0']/100
merged_table['UNIT'] = merged_table['UNIT'].astype(float)
```

```python
# טבלת סיכום לפי ני"ע לדוגמא
table = merged_table[["comp_num","Maslul","sheet","NAME","ISIN",'YEAR','QTR','PV','new_pv','gap#','gap%','OLD_PRICE',"Price","UNIT"]].round(2)
table.sample(10)
```

```python
# תוצאות מאוגרגות
result = table.groupby(["comp_num","Maslul","YEAR","QTR"]).agg({"PV":"sum",'new_pv':"sum",'gap#':"sum",'gap%':"mean",'OLD_PRICE':"sum","Price":"sum","UNIT":"sum"})
result
```
